```python
import pymysql
import pandas as pd
# save 30 into recom1
def insert_recom1(a_list):
    
    db = pymysql.connect(host='mysql', user='user',password='user',
                           database='project',charset="utf8",port=3306)
    cursor = db.cursor()
    df=pd.DataFrame({"srno": [], "jobCode": [], "jobName": [], "jobDescription": [], "C": []})

    # colect 30
    for q in a_list:
        sql_2 = "select srno, jobCode, jobName, jobDescription, C from s104 where srno in ({})".format(q) 
        data = pd.read_sql(sql_2, db) #利用pandas直接获取数据
        df = df.append(data, ignore_index=True)

    # insert 30
    for i in range(0,30):
        C1=df.loc[i]['srno']
        C2=df.loc[i]['jobCode']
        C3=df.loc[i]['jobName']
        C4=df.loc[i]['jobDescription']
        C5=df.loc[i]['C'] 

        sql_1 = "insert into recom_1(srno, jobCode, jobName, jobDescription, C) values( %s,'%s','%s','%s','%s')" % \
                (C1,C2,C3,C4,C5)
        try:
            cursor.execute(sql_1) # 执行sql语句
            db.commit() # 提交到数据库执行
        except:
            db.rollback() # 如果发生错误则回滚


    db.close()
```
